```python
liste=[]
agirlik=[]
agirlikToplam=0
for i in range(3):
    puan=int(input("Puan Giriniz:"))
    liste.append(puan)
    x=float(input("Ağırlığını Giriniz:"))
    agirlikToplam+=x
    agirlik.append(x)
   
toplam=0
for a in range(3):
    toplam+=liste[a]+agirlik[a]

sonuc=toplam/3
print(sonuc)

if sonuc>85:
    print("AA")
elif sonuc>75 :
    print("BA")
elif sonuc>65:
    print("BB")
elif sonuc>55:
    print("CB")
elif sonuc>45:
    print("CC")
else:
    print("Kaldınız.")
    
```

```python
def dersNotunuAl(dersNotları, dersNumarasi):
    while True:
        try:
            ders_notu = int(input(f"{dersNumarasi} için ders notunu giriniz: "))
            if ders_notu < 0 or ders_notu > 100:
                print("Ders notu 0-100 arasında olmalıdır.")
                continue
            break
        except ValueError:
            print("Lütfen bir tam sayı giriniz.")

    while True:
        try:
            agirlik = int(input(f"{dersNumarasi} için ağırlık notunu giriniz: "))
            if agirlik < 0 or agirlik > 10:
                print("Ağırlık notu 0-10 arasında olmalıdır.")
                continue
            break
        except ValueError:
            print("Lütfen bir tam sayı giriniz.")

    dersNotları.append((ders_notu, agirlik))


def agırlıkHesapla(dersNotları):
    toplam = 0
    toplamAgırlık = 0
    for notAgırlık in dersNotları:
        notu = notAgırlık[0]
        agırlık = notAgırlık[1]
        toplam += notu * agırlık
        toplamAgırlık += agırlık

    return toplam / toplamAgırlık


def harfNotuYazdir(gano):
    print("GANO: ", gano)
    print("HARF NOTU: ", end="")
    if gano >= 85:
        print("AA")
    elif gano >= 70:
        print("BA")
    elif gano >= 55:
        print("BB")
    elif gano >= 40:
        print("CB")
    else:
        print("FF")


dersNotları = []

while True:
    try:
        dersSayisi = int(input("Ders sayısını giriniz: "))
        if dersSayisi <= 0:
            print("Ders sayısı pozitif bir tam sayı olmalıdır.")
            continue
        break
    except ValueError:
        print("Lütfen bir tam sayı giriniz.")

for i in range(dersSayisi):
    dersNotunuAl(dersNotları, i+1)

gano = agırlıkHesapla(dersNotları)
harfNotuYazdir(gano)
```
