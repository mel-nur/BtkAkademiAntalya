```python
#Yaşa göre hayatta kalma dağılımı
plt.figure(figsize=(7,7))
sns.histplot(data=df,x='Age',hue='Survived',multiple='stack',kde=True,palette='pastel',bins=10,alpha=.5)
plt.title("Yaş-Hayatta kalma grafiği")
plt.show()
```

```python
sns.barplot(x='Sex',y='Survived',data=df,ci=None,palette=["pink","blue"])
for container in plt.gca().containers:
    labels=[f'{int(value.get_height()*100)}%' for value in container]
    plt.bar_label(container,labels=labels)
plt.xlabel("Cinsiyet")
plt.ylabel("Hayatta kalma oranı")
plt.title("Cinsiyete göre hayatta kalma")
plt.show()
```
